The Column V2 contains only zero. So, we should drop it.

```python
df1 = df.copy()
```

```python
df1.drop('V2', axis=1, inplace=True)
```

```python
col_names = list(df1.columns)
```

```python
for col in col_names:
    sns.boxplot(y=df1[col])
    plt.show()
```

<b> Observations : </b>
- Few columns have Outliers, need to remove.

# Removing Outliers

```python
from scipy import stats
import numpy as np
for col in col_names:
    df1=df1[(np.abs(stats.zscore(df1[col])) < 3)]
```

```python
df1
```

```python
for col in col_names:
    sns.distplot(df1[col])
    plt.show()
```

<b> Observations :</b>
- Even columns are symmetric.
- Odd columns are little skewed, although not long tailed.

```python
plt.figure(figsize=(20,20))
sns.heatmap(df1.corr(),annot=True)
plt.show()
```

```python
df1.dtypes
```

```python
y=df1['Class']
```

```python
X=df1.drop('Class',axis=1)
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_std = ss.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtr=DecisionTreeClassifier(max_depth=5)
dtr.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

#Making predictions
y_pred=dtr.predict(X_test)

#Printing classification report
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

#specify number of folds for k-fold CV
n_folds = 5

#parameters to build model on
parameters={'max_depth':range(5,30)}

#instantiate the model
dtree = DecisionTreeClassifier(criterion='gini',random_state=100)

#fit tree on the training data
tree=GridSearchCV(dtree, parameters, cv=n_folds, scoring='accuracy')

tree.fit(X_train,y_train)
```

```python
tree.best_params_
```

```python
scores=tree.cv_results_
pd.DataFrame(scores).head(20)
df_scores = pd.DataFrame(scores)
```

# USING Entropy

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

#specify number of folds for k-fold CV
n_folds = 5

#parameters to build model on
parameters={'max_depth':range(5,30)}

#instantiate the model
dtree_entr = DecisionTreeClassifier(criterion='entropy',random_state=100)

#fit tree on the training data
tree_entr=GridSearchCV(dtree_entr, parameters, cv=n_folds, scoring='accuracy')

tree_entr.fit(X_train,y_train)
```

```python
tree_entr.best_params_
```

```python
scores_entr=tree_entr.cv_results_
pd.DataFrame(scores_entr).head(20)
df_scores_entr = pd.DataFrame(scores_entr)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

plt.plot(df_scores['param_max_depth'], df_scores['mean_test_score'], color = 'red', linewidth = 2)
plt.plot(df_scores_entr['param_max_depth'], df_scores_entr['mean_test_score'], color = 'green',linewidth=2 )
#plt.errorbar(df_scores['param_max_depth'], df_scores['mean_test_score'], df_scores['std_test_score'],color = 'red', linewidth = 2)
#plt.errorbar(df_scores_entropy['param_max_depth'], df_scores_entropy['mean_test_score'], df_scores_entropy['std_test_score'] ,color = 'green',linewidth=2 )
plt.xlabel('Max_depth')
plt.xlim(3,30)
#plt.ylim(0.805, 0.82)
plt.ylabel('Mean test score')
plt.show()
```

<b>Observations : </b> 
- Gini index method is performing better than the Entropy method.
- So, from now I will use Gini index method for final model.

# Tuned Model

```python
model_tree = DecisionTreeClassifier(criterion='gini', max_depth=9)
model_tree.fit(X_train,y_train)
#Making FINAL predictions
y_pred_final=model_tree.predict(X_test)

#Printing classification report
print(classification_report(y_test,y_pred_final))
```

```python
print(confusion_matrix(y_test,y_pred_final))
```

```python
pred_prob = model_tree.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve

# roc curve for models
fpr, tpr, thresh = roc_curve(y_test, pred_prob[:,1], pos_label=1)


# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.plot(p_fpr, p_tpr, marker='.', label='')
plt.plot(fpr, tpr, marker='o', label='Decision Tree')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

# auc scores
auc_score = roc_auc_score(y_test, pred_prob[:,1])
print(auc_score)
```
